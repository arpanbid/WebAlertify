```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

import time
import requests, openpyxl
import pandas as pd
import re
```

```python
try:
    xl_data = pd.read_excel('DateCheck.xlsx', sheet_name="Sheet1")
except Exception as e:
    print("Errror in reading excel file.\n" + e)
    
```

```python
xl_old_data = xl_data.copy()
xl_data
```

```python
#browser=webdriver.Chrome('/Users/arpanbid/WebDriver/chromedriver') #depreciated method

browser = webdriver.Chrome(service=Service('/Users/arpanbid/WebDriver/chromedriver'))
```

```python

try:
    for i in range(0,len(xl_data)):


        link = xl_data.Link[i]
        name = xl_data.Name[i]
        #tag = xl_data.tag[i]
        currentQ = xl_data.CurrentQ[i]
        lastQ = xl_data.LastQ[i]




        browser.get(link)
        time.sleep(5)
        soup = BeautifulSoup(browser.page_source)



        p = soup.find_all(string= re.compile(lastQ))
        c = soup.find_all(string= re.compile(currentQ))

        if p == []:
            xl_data.at[i,'Result']="Error"
            print(name, ": Error")
        else:
            if c == []:
                print(name, ": Negative")
                xl_data.at[i,'Result']="Negative"
            else:
                print(name, ": Positive")
                xl_data.at[i,'Result']="Positive"


except Exception as e:
    print("Error in fetching data from " + link + "\n\n" + e)
    
    
browser.quit()
```
