```python
from bs4 import BeautifulSoup
import requests, openpyxl
import pandas as pd
import re
from time import sleep
from selenium import webdriver 
```

```python
try:
    xl_data = pd.read_excel('DateCheck.xlsx', sheet_name="Sheet1")
except Exception as e:
    print("Errror in reading excel file.\n" + str(e))
```

```python
xl_old_data = xl_data.copy()
xl_data
```

```python
edge_options = webdriver.EdgeOptions()
edge_options.add_argument('headless')
driver = webdriver.Edge(options=edge_options)
```

```python
for i in range(0,len(xl_data)):
#for i in range(10,11):

    current_result = xl_data.Result[i]
    link = xl_data.Link[i]
    name = xl_data.Name[i]
    #tag = xl_data.tag[i]
    currentQ = str(xl_data.CurrentQ[i])
    lastQ = str(xl_data.LastQ[i])

    if (((current_result == "Negative") or  (current_result == "Error")) and (currentQ != 'nan')):
        try:
            #headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
            driver.get(link)
            sleep(5)
            soup = BeautifulSoup(driver.page_source,'html.parser')

            p = soup.find_all(string= re.compile(str(lastQ)))
            c = soup.find_all(string= re.compile(currentQ))

            if p == [] and lastQ != "nan":
                xl_data.at[i,'Result']="Error"
                print(name, ": Error")
            else:
                if c == []:
                    print(name, ": Negative")
                    xl_data.at[i,'Result']="Negative"
                else:
                    print(name, ": Positive")
                    xl_data.at[i,'Result']="Positive"
        except Exception as e:
            print("Error in fetching data from " + link + "\n\n" + str(e))
```

```python
driver.quit()
```
